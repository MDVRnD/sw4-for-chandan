```python
from obspy import read
```

```python
for comp in ("mxx", "myy", "mzz"):
    for chan in ("x", "y", "z"):
        # print(comp)
        stream = read("SW4/Dirac/"+comp+".dir/rec50."+chan, format="SAC")
        b = stream[0].stats.starttime; # ti = b+5; tf = b+35;
        # stream.filter('lowpass', freq=5.0)
        stream.filter('bandpass', freqmin=1, freqmax=5)
        # print(comp)
        for tr in stream:
            tr.stats.station = comp
        stream.plot(starttime=b+6, endtime=b+36, type='relative');
```
